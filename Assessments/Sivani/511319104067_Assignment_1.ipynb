## 7. Concatinate a and b 
## a = np.array([1, 2, 3]), b = np.array([4, 5, 6])

```python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
array=np.concatenate((a,b))
print(array)
```
